```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
print(x + y)   
print(x - y)    
print(x * y)  
print(x / y)    
```

```python
a = np.array([0, np.pi/2, np.pi])
print(np.sin(a))
b = np.array([0, 1, 2, 3])
print(np.exp(b))
print(np.sqrt(b))
print (np.sqrt(a))
```

```python
dtype = [('name', 'S10'), ('year', int), ('cgpa', float)]
vals  = [('Hrithik', 2009, 8.5),
         ('Ajay',    2008, 8.7),
         ('Pankaj',  2008, 7.9),
         ('Aakash',  2009, 9.0)]
a = np.array(vals, dtype=dtype)
print(np.sort(a, order='name'))
print(np.sort(a, order=['year', 'cgpa']))
```
